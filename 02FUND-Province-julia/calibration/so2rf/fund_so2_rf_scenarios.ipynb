# FUND SO2 scenarios

## I. Introduction

The SO2 radiative forcing scenarios are exogenous. Historical forcings are based on numbers from RCP85 up until 2005. After that time the FUND scenarios is based on the RCP85 scenario, and the SRES scenarios on the IMAGE 2.2 interpretation of these scenarios.

## II. Data sources

### 1) IMAGE 2.2

This data was obtained from http://themasites.pbl.nl/tridion/en/themasites/image/projects/downloads/index-2.html on 11/18/2013 by Anthoff.

Steps:

- Download and install the User Support System (USS)
- Run Start_SRES_viewer.bat in the install directory
- Using Tools->Scenario Manager, load all the relevant scenarios, then select them one by one to export the data
- Select Aggregated Indicators (IPCC Regions) View -> State -> Other State Indicators -> Climate -> Relvative Contributions to Radiative Forcing -> Negative Contributors
- Right click the area around the graph and select "Save selection as Excel..." and pick a filename
- Open each of the saved files with a newer version of Excel and save it as an xlsx file

```python
image22_scenarios = ['A1b', 'A2', 'B1', 'B2']
data = {}

for image22_scenario in image22_scenarios:
    temp_df = pandas.read_excel('main.DELTQNEG_' + image22_scenario + '.xlsx', 'main.DELTQNEG')  
    temp_df['t'] = temp_df['t'].astype(int)
    # Filter out non sulphur forcings
    temp_df = temp_df[(temp_df['Contributor (a)']=='Sulph. direct') | (temp_df['Contributor (a)']=='Sulph. indirect')]
    
    # Aggregate direct and indirect forcings
    temp_df = temp_df['Negative~Contributors'].groupby(temp_df['t']).sum()
    
    # Negate sign so that forcing is negative
    data['IMAGE 2.2 SRES ' + image22_scenario] = -temp_df
```

```python
df_image22 = pandas.DataFrame(data)
```

```python
df_image22.loc[2005:2100].plot()
xlabel('year')
ylabel('W/m^2')
```

### 2) RCP

Data obtained from http://www.pik-potsdam.de/~mmalte/rcps/ on 11/15/2013 by Anthoff.

```python
temp_df = pandas.read_excel('RCP85_MIDYEAR_RADFORCING.xls', 'RCP85_MIDYEAR_RADFORCING',
                            skiprows=59)
temp_df.index = temp_df['v YEARS/GAS >'].astype(int)
temp_df.index.name = 'year'
```

```python
df_rcp = pandas.Series(temp_df['TOTAER_DIR_RF'] + temp_df['CLOUD_TOT_RF'], name='RCP85')
```

```python
df_rcp.loc[1950:2500].plot(figsize=(10,10),label='RCP85')
xlabel('year')
ylabel('W/m^2')
legend()
```

## III. FUND scenarios

Here we construct scenarios for FUND.

```python
df_fund = pandas.DataFrame(index=range(1950,3001))
```

### 1) FUND scenario

This constructs an emission scenario that has emissions as RCP85 from 1950-2500 and then stays constant at the 2500 emission level for the rest of the time.

```python
new = df_rcp.reindex(range(1950,3001)).interpolate()
df_fund['Base'] = new
```

### 2) SRES scenarios

This constructs emission scenarios that have historical emissions as in RCP85 from 1950-2005, followed by emissions as in the IMAGE 2.2 SRES scenarios for the years 2006-2100, and then stays constant at the 2100 emission level for the rest of the time.

```python
switch_year = 2006
```

```python
for scenario in image22_scenarios:
    new = df_image22['IMAGE 2.2 SRES %s'%scenario].reindex(range(1950,3001))
    new.ix[1950:switch_year-1] = df_rcp.ix[1950:switch_year-1]
    new = new.interpolate()
    df_fund['SRES %s'%scenario] = new
```

### 3) Figures

```python
df_fund.plot(figsize=(10,10))
xlabel('year')
ylabel('W/m^2')
```

```python
df_fund.loc[2000:2020].plot(figsize=(10,10))
xlabel('year')
ylabel('W/m^2')
```

### 3) Output

Write the input file for the model.

```python
for scen in df_fund.columns:
    with open('../../Fund/Data/%s/rfso2.csv' % scen, 'w') as f:
        f.write('# %s scenario\n' % scen)            
        f.write('# Generated by calibration/so2rf/fund_so2_rf_scenarios.ipynb\n')            
        df_fund[scen].to_csv(f)
```
